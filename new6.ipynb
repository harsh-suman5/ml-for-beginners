```python
import pandas as pd
import sklearn as sns
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('ADANIPORTS_minute.csv')
data.head()
```

```python
data.describe()
```

```python
#data loading and preparation

from sklearn.model_selection import train_test_split

x = np.array(data['close']).reshape(-1,1)
y = np.array(data['open']).reshape(-1,1)
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,shuffle=False)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(x_train,y_train) 
```

```python
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

y_pred = reg.predict(x_test)

print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred):.2f}")
print(f"R2 Score: {r2_score(y_test, y_pred):.2f}")

# Training data + regression line

plt.scatter(x_train.ravel(), y_train, color= "black", label = "train data")
plt.plot(x_train,reg.predict(x_train),color = "blue",linewidth=2, label ="model")
plt.title("training data")
plt.xlabel("close")

plt.ylabel("open")
plt.legend()
plt.show()
# Testing data + regression line
plt.scatter(x_test.ravel(), y_test, color="orange", label="Test data")
plt.plot(x_test, y_pred, color="red", linewidth=2, label="Prediction")
plt.title("Testing Data")
plt.xlabel("close")
plt.ylabel("open")
plt.legend()
plt.show()
```
