```python
#simple linear regression plots
import pandas as pd
import matplotlib.pyplot as plt
data  = pd.DataFrame({"cigrates":[0,10,20,30],
                     "CVD": [572,802,892,1025],
"lung": [14,105,208,355]})
plt.ylabel("CVD deaths")
data.plot("cigrates", "CVD", kind="scatter")
plt.title("Deaths for different smoking intensities")
plt.xlabel("Cigarettes smoked per day")
```

```python
# fitting a linear model
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
df = pd.DataFrame({"cigrates":[0,10,20,30],
                     "CVD": [572,802,892,1025],
"lung": [14,105,208,355]})    
df.plot("cigrates","CVD", kind="scatter")
lm = smf.ols("CVD ~ cigrates", data=df).fit()
xmin = df.cigrates.min()
xmax = df.cigrates.max()
x = np.linspace(xmin, xmax, 100)
y = lm.params[0] + lm.params[1] +x
plt.plot(x,y, color="darkgreen")
```
