```python
import pandas as pd
data = pd.read_csv("CCPP/Folds5x2_pp.csv")
data.head()
```

```python
data.describe()
```

```python
# visualising the multivalued datasets
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

scatter_matrix(data, diagonal="kde")
```

```python
#3d plotting

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as  plt 
fig = plt.Figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data["AT"],
data["V"], data["PE"]
)
ax.set_xlabel("AT")
ax.set_ylabel("V")
ax.set_zlabel("PE")
plt.show()
```

```python
import statsmodels.formula.api as smf
lm = smf.ols(formula='PE~AT+V+AP+RH', data=data).fit()
print(lm.params) 
```

```python
m = pd.DataFrame({
    "AT":[9.45],
    "V":[10.5],
    "AP":[12.34],
    "RH":[8.7]
})
lm.predict(m)
```
